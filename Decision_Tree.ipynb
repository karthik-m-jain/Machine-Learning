```python
#import librariees
import pydotplus
import numpy as np
import pandas as pd

#Read dataset
df=pd.read_csv("C:\\Users\\admin\\Desktop\\CS\\ML\\dt.csv")

df
```

```python
#Distribute the features
x=df.iloc[:,:-1]
feature_cols=['age','income','student','credit_rating']
y=df.iloc[:,4].values
```

```python
#Encode the class label
from sklearn.preprocessing import LabelEncoder
label_encoder_X=LabelEncoder()
x=x.apply(LabelEncoder().fit_transform)
```

```python
#Import Decision tree classifier 
from sklearn.tree import DecisionTreeClassifier

#Using Entropy criterion 
regressor = DecisionTreeClassifier(criterion='entropy',max_depth=None,splitter='best')
feature_cols
```

```python
#Fit the model
regressor.fit(x,y)

#Predict for random features
x_in=np.array([1,1,0,0])
y_pred=regressor.predict([x_in])
print("Prediction:",y_pred)
```

```python
#Import libraries for decision tree generation
from sklearn.externals.six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus
```

```python
#Store decision tree as PNG image
#export_graphviz function converts decision tree classifier into dot file and pydotplus convert this dot file to png or displayable form on Jupyter.

dot_data=StringIO()
export_graphviz(regressor,out_file=dot_data,filled = True,rounded=True,special_characters=True,feature_names=feature_cols,class_names=['0','1'])
graph=pydotplus.graph_from_dot_data(dot_data.getvalue())

graph.write_png('DT.png')
```

```python
#Diabetes dataset
df=pd.read_csv("C:\\Users\\admin\\Desktop\\CS\\ML\\pima-indians-diabetes.csv")
```

```python
#First 5 rows of the dataset
df.head()
```

```python
#Categorize the columns
x=df.iloc[:,:-1].values
feature_cols=['pregnant','glucose','bp','insulin','bmi','pedigree','age','label']
y=df.iloc[:,-1].values
```

```python
#split the data into train and test

from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.3,random_state=1)

classifier=DecisionTreeClassifier(criterion='entropy',max_depth=None,splitter='best')
```

```python
#Fit the model
classifier.fit(X_train,Y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
#Print the predicted and actual values
df1 = pd.DataFrame({'Actual': Y_test.flatten(), 'Predicted': y_pred.flatten()})          #Comparing predicted and actual data
df1
```

```python
#Determine the accuracy of Entropy criterion
from sklearn import metrics

print("Accuracy : ",metrics.accuracy_score(Y_test,y_pred))
```

```python
#Generate the PNG image

dot_data=StringIO()
export_graphviz(classifier,out_file=dot_data,filled = True,rounded=True,special_characters=True,feature_names=feature_cols,class_names=['0','1'])
graph=pydotplus.graph_from_dot_data(dot_data.getvalue())

graph.write_png('diabetes_entropy.png')
```

```python
#Classification using GINI INDEX criterion
classifier=DecisionTreeClassifier(criterion='gini',max_depth=None,splitter='best')

classifier.fit(X_train,Y_train)

y_pred=classifier.predict(X_test)
```

```python
#Accuracy
print("Accuracy",metrics.accuracy_score(Y_test,y_pred))
```

```python
#Entropy crtierion has more accuracy than Gini Index criterion
```

```python
dot_data=StringIO()
export_graphviz(classifier,out_file=dot_data,filled = True,rounded=True,special_characters=True,feature_names=feature_cols,class_names=['0','1'])
graph=pydotplus.graph_from_dot_data(dot_data.getvalue())

graph.write_png('diabetes_gini.png')
```
